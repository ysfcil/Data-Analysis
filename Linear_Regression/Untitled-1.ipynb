```python
def psi(x_, k):
    y0 = []
    for i in range(len(x_)):
        if k == 0:
            y0.append(x_[i])
        else:
            y0_ = np.exp(1j*k*x_[i]-2*abs(k))/1j*k
            y0.append(y0_)
    return y0
```

```python
x0 = np.linspace(-5, 5)

y3 = psi(x0, 0.5)
y4 = psi(x0, 1)
y5 = psi(x0, 2)

plt.plot(x0, y3, label = "k=0.5")
plt.plot(x0, y4, label = "k=1")
plt.plot(x0, y5, label = "k=2")
plt.legend()

```
