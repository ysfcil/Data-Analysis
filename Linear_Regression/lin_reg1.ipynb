```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
```

```python
x = np.linspace(1, 100, 99)
y = np.zeros(99)
for i in range(len(x)):
    y[i] = 3*x[i]+25+np.random.randint(-40, 40)
plt.scatter(x, y)
plt.plot(x, 3*x+25, color = "C1")
```

```python
import sklearn.linear_model
regr = sklearn.linear_model.LinearRegression()
x = x[:, np.newaxis]
regr.fit(x, y)
```

```python
y_hat = regr.predict(x)
```

```python
plt.scatter(x, y)
plt.plot(x, 3*x+25, color = "C1")
plt.plot(x, y_hat, color = "green")
```
