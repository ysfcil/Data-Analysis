```python
import pandas as pd
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table(r"C:\Users\yusuf\Downloads\users.dat", sep='::', header=None, names=unames, engine="python")
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table(r"C:\Users\yusuf\Downloads\ratings.dat", sep='::', header=None, names=rnames, engine="python")
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table(r"C:\Users\yusuf\Downloads\movies.dat", sep="::", header=None, names=mnames, engine="python")
```

```python
genre_count = movies["genres"].value_counts()
genre_count[:10]
gender_count = users["gender"].value_counts()
gender_count
```

```python
df = pd.merge(pd.merge(ratings, users), movies)
genre_group = df.groupby(["age", "genres"]).rating.mean()
genre_group







```

```python
mean_ratings = df.pivot_table("rating", index = ["title"], columns= "gender", aggfunc="mean")
female_rating_sorted = mean_ratings.sort_values(by=["F"], ascending=False)
female_rating_sorted[:20].plot(kind="barh")
```

```python
genre_gender = df.pivot_table("rating", index = ["genres"], columns= "gender", aggfunc="mean")
genre_gender=genre_gender.sort_values(by="F", ascending=False)
genre_gender.unstack()
genre_gender[:15].plot(kind="barh")#Women's favourite genres compared with men plotted
```
